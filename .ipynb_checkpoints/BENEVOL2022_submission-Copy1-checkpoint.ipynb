```python
import code_extract as ce
import pandas as pd
import numpy as np
```

```python
import spacy

nlp = spacy.load('en_core_web_md')
nlp.max_length = 4000000
```

```python
ml_projects = ["angel",
               "smile",
               "vespa",
               "Alink",
               "serve",
               "djl",
               "EasyML",
               "oryx",
               "grobid",
               "seldon-server"]

api_projects = ["openapi-generator",
                "spring-boot-api-project-seed",
                "android-hidden-api",
                "JDA",
                "logging-log4j2",
                "Component",
                "swagger2markup",
                "CameraFragment",
                "docker-java",
                "ApiManager"]

sec_projects = ["zaproxy",
                "cryptomator",
                "spring-security",
                "graylog2-server",
                "DependencyCheck",
                "hawk",
                "MifareClassicTool",
                "pac4j",
                "jasypt-spring-boot",
                "Note"]

fe_projects = [ "AmazeFileManager",
                "spring-boot-vuejs",
                "Jailer",
                "jExifToolGUI",
                "kanban-board",
                "openseedbox",
                "frontend-gradle-plugin",
                "aem-core-cif-components",
                "acai",
                "design-library-plugin"]


dv_projects = ["dataease",
                "poli",
                "AnyChart-Android",
                "jfreechart",
                "AAChartCore",
                "abixen-platform",
                "datagear",
                "Knowage-Server",
                "helicalinsight",
                "intermine"]
```

```python
topics = ["machine-learning", "api", "security", "front-end", "data-visualization"]
topics = [[x]*10 for x in topics]
topics = [item for sublist in topics for item in sublist]
```

```python
all_projects = ml_projects + api_projects + sec_projects + fe_projects + dv_projects

projects_df = pd.DataFrame(all_projects, columns=["Projects"])

projects_df['Topic'] = topics
```

```python
projects_df
```

```python
%%time
code_data_extracted = []
ctr = 0

for proj in all_projects:
    code_data_extracted.append(ce.extract_from_code('repo2\\{}\\'.format(proj), topics[ctr], nlp))
    print("{} extracted...".format(proj))
    ctr+=1
```

```python
df = pd.concat(code_data_extracted)
df = df.dropna()
df = df.drop(df[df['Data']==''].index)
df = df.reset_index(drop=True)
df
```

```python
empty_rows = df[df['Data']=='']
empty_rows
```
